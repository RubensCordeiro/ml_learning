<h1> On this notebook, I will explore with some models to create a number 6 detector </h1>

```python
import pandas as pd
import numpy as np
import helper 
```

```python
#Data Loading
folder_path = "datasets"
data_file = "train.csv"

#Binary datasets
binary_train_set = helper.data_loader(folder_path, data_file, 'train', 'binary', 6)
binary_test_set = helper.data_loader(folder_path, data_file, 'test', 'binary', 6)

'''
#Multilabel datasets
multi_train_set = helper.data_loader(folder_path, data_file, 'train', 'multinomial')
multi_test_set = helper.data_loader(folder_path, data_file, 'test', 'multinomial')

#Multioutput
multil_train_set = helper.data_loader(folder_path, data_file, 'train', 'multilabel')
multil_test_set = helper.data_loader(folder_path, data_file, 'test', 'multilabel')
'''
```

```python
x_train, y_train = binary_train_set
x_prepared = helper.data_pipeline.fit_transform(x_train)
```

```python
#Implementing a KNN model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, cross_val_predict

knn_model = KNeighborsClassifier()

scores = cross_val_score(knn_model, X = x_prepared, y = y_train, scoring = 'accuracy', cv = 5)
scores
```

```python
#Mean and std values
print(scores.mean(), scores.std())
```

```python
#Running a few models to search for good candidates
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier

models = {
    'KNN' : KNeighborsClassifier(),
    'Naive Bayes' : MultinomialNB(),
    'SGD Classifier' : SGDClassifier(),
    'Random Forest' : RandomForestClassifier()
}

parameters = {
    'KNN' : {'weights' : ['uniform', 'distance'], 'n_neighbors' : [3,5,7]},
    'Naive Bayes' : {'alpha' : [1]},
    'SGD Classifier' : [{'penalty': ['l2', 'l1', 'elasticnet'], 'fit_intercept': [False, True]}],
    'Random Forest' : {'n_estimators': [1, 10, 50, 100, 200]}
}
```

```python
model_selection = helper.multiModelSelection(models, parameters)
model_selection.fit(x_prepared, y_train, cv = 5, n_jobs = 2, scoring = 'accuracy')
```

```python
model_selection.display_results()
```

```python
#Selecting a few models to test further metrics

rforrest_classifier = RandomForestClassifier(n_estimators = 200).fit(x_prepared, y_train)
knn_classifier = KNeighborsClassifier(n_neighbors = 7).fit(x_prepared, y_train)
sgd = SGDClassifier(fit_intercept = False)
```

```python
import joblib
joblib.dump(rforrest_classifier, 'binary_random_forrest.pkl')
joblib.dump(knn_classifier, 'binary_knn.pkl')
joblib.dump(sgd, 'binary_sgd.pkl')
```
